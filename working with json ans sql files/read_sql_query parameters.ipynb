<h2 style="color:brown">Create SQLAlchemy engine (THIS replaces conn)</h2>

```python
engine = create_engine("mysql+mysqlconnector://root:@localhost/world")
```

<h1 style="color:red">Parameters</h1>

<h2 style="color:brown">index_col Parameter</h2>

```python
# index_col :-  str or list of str, optional, default: None
# Column(s) to set as index(MultiIndex).

df6 = pd.read_sql_query("SELECT * FROM city", engine, index_col='ID')
df6.info()
df6
```

<h2 style="color:brown">params Parameter</h2>

```python
# To safely pass values into SQL queries  (prevents SQL injection)

# BAD way ❌ (Never do this)
# city = "Delhi"
# sql = "SELECT * FROM city WHERE name = '{city}'"

# GOOD way ✅ (params)
sql = "SELECT * FROM city WHERE CountryCode = %s"
df7 = pd.read_sql_query(sql, engine, params=("IND",))
display(df7)

# Dictionary-style params
sql2 = "SELECT * FROM city WHERE CountryCode = %(cntCode)s AND Population > %(popCnt)s ORDER by Population DESC"
df7 = pd.read_sql_query(sql2, engine, params={"cntCode":"IND", 'popCnt':'1000000'})
display(df7)
```

<h2 style="color:brown">chunksize Parameter</h2>

```python
data = pd.read_sql_query(      # Instead of returning one big DataFrame, it returns an iterator of smaller DataFrames.
    "SELECT * FROM city",
    engine,
    chunksize=1000
)

for chunk in data:
    print(chunk.shape)
```

<h2 style="color:brown">dtype Parameter</h2>

```python
df9 = pd.read_sql_query(      # Instead of returning one big DataFrame, it returns an iterator of smaller DataFrames.
    "SELECT * FROM country",
    engine,
    dtype={"GNP":int}
)

df9.info()
display(df9)
```
