<h2>date_unit parameter</h2>

```python
# date_unit
# Type: str | None
# Default: None
# Specifies the unit of numeric timestamps when converting to datetime.

# Allowed Values : 's', 'ms', 'us', 'ns'

df2 = pd.read_json("https://api.exchangerate-api.com/v4/latest/INR", date_unit="ns")
df2.info()
df2
```

<h2> convert_dates parameter</h2>

```python
#convert_dates:  bool or list of str, default true
# True	:  Convert default date-like columns
# False :  Do NOT convert any date columns
# ["col1", "col2"] :	Convert only these columns (plus defaults if allowed)

df2 = pd.read_json("https://api.exchangerate-api.com/v4/latest/INR", convert_dates=False)
df2.info()
display(df2)

pd.read_json("https://api.exchangerate-api.com/v4/latest/INR", keep_default_dates=False, convert_dates=["date"]).info()

# Default “date-like” column names              : for -> True
# A column is automatically considered date-like if its name:
# ends with _at → created_at
# ends with _time → start_time
# starts with timestamp
# is named modified
# is named date
```

<h2>keep_default_dates parameter</h2>

```python
# keep_default_dates : bool
# Default: True
# Controls whether pandas also converts default date-like columns when convert_dates is enabled.

# True	Convert default date-like columns
# False	Convert ONLY columns listed in convert_dates
pd.read_json("https://api.exchangerate-api.com/v4/latest/INR", keep_default_dates=False).info()
```

<h2>typ parameter</h2>

```python
# typ -->  {‘frame’, ‘series’}, default ‘frame’
# typ tells pandas what kind of object to return after reading JSON data.
                                                                              #   series               #  frame
# typ = "frame"   → returns a DataFrame (DEFAULT)                                {                     {
# typ = "series"  → returns a Series                                               "x": 10,                 "A": [1, 2, 3],
                                                                        #          "y": 20,                 "B": [4, 5, 6]
# frame -> Suitable for tabular data analysis                           #          "z": 30             }
df4 = pd.read_json("train.json", typ="frame")                           #        }
display(df4)                                                           
# series -> Useful when JSON represents key–value pairs
df4 = pd.read_json("train.json", typ="series")            
display(df4)
```

<h2>orient parameter</h2>

```python
# orient   -> str, optional
# orient tells pandas HOW the JSON data is organized
# i.e., how rows, columns, and values are stored inside the JSON file.
# Pandas does not convert JSON structure — it only interprets it using orient.
    
# Indication of expected JSON string format. Compatible JSON strings can be produced by to_json() with a corresponding orient value.
# The set of possible orients is:                                                             # split JSON format                    #   records JSON format
# 'split' : dict like {index -> [index], columns -> [columns], data -> [values]}               {                                      [
# 'records' : list like [{column -> value}, ... , {column -> value}]                               "columns": ["id", "name"],            {"id":1,"name":"A"},
# 'index' : dict like {index -> {column -> value}}                                                 "index": [0, 1],                      {"id":2,"name":"B"}
# 'columns' : dict like {column -> {index -> value}}                                               "data": [[1, "A"], [2, "B"]]       ]
# 'values' : just the values array                                                             }
# 'table' : dict like {'schema': {schema}, 'data': {data}

# The allowed and default values depend on the value of the typ parameter.
# when typ == 'series',
# allowed orients are {'split','records','index'}
# default is 'index'
# The Series index must be unique for orient 'index'.

# when typ == 'frame',
# allowed orients are {'split','records','index', 'columns','values', 'table'}
# default is 'columns'
# The DataFrame index must be unique for orients 'index' and 'columns'.
# The DataFrame columns must be unique for orients 'index', 'columns', and 'records'.

df5 = pd.read_json('train.json', orient="column")
display(df5)
df5 = pd.read_json('train.json', orient="records")
display(df5)
# split -> ❌ Cannot work on list-based JSON
```
