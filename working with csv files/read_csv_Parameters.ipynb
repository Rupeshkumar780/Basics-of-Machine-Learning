<h1 style='color:red'> 4. Adding Column Names </h1>
<pre><b> if column names are not specified in the data frame, you can add the column names of your choice using <i style='background-color:yellow'>names</i> parameter<b></pre>

```python
df4 = pd.read_csv('web_traffic.tsv', sep='\t', names=['S.no', 'Web Traffic'])
print(df4.columns)   # this line prints the list of column names from the DataFrame
display(df4)
```

<h1 style='color:red'> 5. index_col Parameter </h1>

```python
# df5 = pd.read_csv('web_traffic.tsv', sep='\t', index_col='S.no')      **** Error for the web_traffic.tsv file as it doesn't have header names ****
# If the data already HAS a header, then you can use index_col='S.no' :

#If the file does NOT have a header:    Use both names and index_col together:
df5 = pd.read_csv('web_traffic.tsv', sep='\t', names=['S.no', 'Web Traffic'], index_col='S.no')
# S.no will be now considered the Index of the DataFrame.
print(df5.columns)    # this line prints the list of column names from the DataFrame, we can see the 'S.no' is not appearing as 
display(df5)
```

<h1 style='color:red'> 8. skiprows/nrows parameter </h1>

```python
# df8 = pd.read_csv('heart.csv', skiprows=[2,3,5], header=1)      # skips 2,3,5 th rows ( actual data starts from 2nd row, when header=1 used)
# df8 = pd.read_csv('heart.csv', header=1, skiprows=[0,1])      #  skips the ignored row 0 and the header row 1 (column name row)
# df8 = pd.read_csv('heart.csv', header=1, skiprows=[x for x in range(1000) if x%5 == 0])  

df8 = pd.read_csv('heart.csv', header=1, skiprows=[x for x in range(1000) if x>100])  
        # That means it will skip all rows with index greater than 100, i.e. rows 101 to 999.
        # So:  
        #    skiprows in pandas.read_csv() means:     “Do NOT read these row numbers when the given condition is true.”
        #    So whatever list you pass in skiprows=[ ... ] → those rows will be skipped (ignored), not read.
        # Rows 0–100 → will be read
        # Rows 101–999 → skipped
display(df8)

# df8 = pd.read_csv('heart.csv', nrows=200)   # read the first 200 data rows 
df8 = pd.read_csv('heart.csv', header=1, nrows=200)   # read the first 200 data rows after the header row
display(df8)
```

<h1 style="color:red"> 10. Skip Bad Lines </h1>

```python
# if our dataset have 5 columns, but some rows have some different no. of columns, in that case it will skip that all lines.

df10 = pd.read_csv("heart.csv", on_bad_lines='error')    # 'error': The default behavior. It raises an exception when a bad line is encountered, stopping the process.
df10 = pd.read_csv("heart.csv", on_bad_lines='warn')    #'warn': Raises a warning for the bad line and skips that specific line, continuing to load the rest of the data.
df10 = pd.read_csv("heart.csv", on_bad_lines='skip')    #'skip': Silently skips the bad lines without raising any error or warning.

display(df10)
```

<h1 style="color:red">11. dtype Parameter</h1>

```python
df11 = pd.read_csv("patients.csv", dtype={"age":float, "service":object})     # changes the data type of 'age' column: int -> float
display(df11)
df11.info()         # complete information about everyrow of dataset
```

<h1 style="color:red">12. Handling Dates</h1>

```python
# in dataset, Dates are stored in object(sting) format, so we can't apply any filters in this columns  as year > 2000
df12 = pd.read_csv("patients.csv", parse_dates=['arrival_date','departure_date'])    # arrival_date, departure_date data types = datetime64
# now we can apply filters in date for 'arrival_date','departure_date' after parsing them into dates, 

# if any two three columns have the date information, then it can be parsed as a single column.   eg :- parse_dates=[['arrival_month',arrival_year'], 'departure_date']
# list of list ['arrival_month',arrival_year']considered as a single column

# we can even pass a dict  parse_dates={"date":['arrival_month',arrival_year']}

display(df12)
df12.info()


# parse_dates: bool or list of int or names or list of lists or dict, default False

# boolean. If True -> try parsing the index.
# list of int or names. e.g. If [1, 2, 3] -> try parsing columns 1, 2, 3 each as a separate date column.
# list of lists. e.g. If [[1, 3]] -> combine columns 1 and 3 and parse as a single date column.
# dict, e.g. ['foo': [1,3]-> parse columns 1, 3 as date and call result 'foo
```

<h1 style="color:red">13. Converters</h1>


```python
def rename(name):                  # a function named 'rename' is created, and 'name' is taken as a parameter
    if name == "Julia Torres":
        return "JT"                    
    else: 
        return name

#  in name column of patient.csv file, change (Julia Torres -->  JT) where ever you will find
df13 = pd.read_csv('patients.csv', converters={'name':rename})    # dictionary ---> {'column_name':function_name}
display(df13)
```

<h1 style="color:red">14. na_values Parameter</h1>

```python
df14 = pd.read_csv("DS.csv", sep=';', nrows=5, encoding="latin-1")   
display(df14)
df14 = pd.read_csv("DS.csv", sep=';', nrows=5, encoding="latin-1", na_values=['_',"?",'not available'])    # replace all the given values with NaN one
display(df14)
```
